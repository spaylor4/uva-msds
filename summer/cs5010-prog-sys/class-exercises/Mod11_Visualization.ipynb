```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
plt.rc('figure', figsize=(10, 6)) #change default plot output size
```

```python
#read in data
#using some golf data I downloaded for an old project
golf = pd.read_csv("/Users/shannon/Documents/code/strokes_gained/SGTee-to-Green.csv")
```

```python
golf.info()
```

```python
subset = golf[golf['PLAYER_NAME'].isin(['Sergio Garcia', 'Ernie Els', 'Tiger Woods', 
                                        'Phil Mickelson', 'Justin Rose'])]
```

### Line Plot

```python
sns.lineplot(x="YEAR", y="SG_OTT", data=subset, hue = "PLAYER_NAME")
```

### Bar Plot

```python
sns.barplot(x='PLAYER_NAME', y='MEASURED_ROUNDS', data=subset[subset['YEAR']==2005])
```

### Histogram

```python
sns.distplot(golf['AVERAGE'], bins=30, color='r')
```

### Scatter Plot

```python
sns.regplot('SG_OTT', 'SG_ARG', data=golf[golf['YEAR']==2018])  # m1 vs unemp columns to plot
plt.title('Strokes Gaines Off Tee vs. Around Green')
```

### Box Plot

```python
sns.boxplot(x='YEAR', y='SG_APR', data=golf)
```
